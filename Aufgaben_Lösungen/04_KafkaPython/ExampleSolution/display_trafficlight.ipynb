```python
import kafka_helpers
import json
c = kafka_helpers.Consumer({'bootstrap.servers': kafka_helpers.kafka_brokers, 'group.id': 'display_group1110', 'auto.offset.reset': 'earliest'})
```

```python
c.subscribe([kafka_helpers.filtered_topic])
import ipywidgets as widgets
%matplotlib inline
import matplotlib.pyplot as plt
from IPython import display

out = widgets.Output(layout={'border': '1px solid black'})
display.display(out)


        
while True:
    msg = c.poll(1.0)

    if msg is None:
        continue
        
    if msg.error():
        print("Consumer error: {}".format(msg.error()))
        continue
    
    kafka_helpers.read_and_display(msg, display)
    
    
c.close()
```
